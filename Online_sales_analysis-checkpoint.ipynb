```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\Hello\OneDrive\Documents\Smart_phone.csv') 
```

```python
print(df.isnull().sum())
```

```python
df_cleaned = df.dropna()
```

```python
df.fillna(0, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.head())
print(df.columns)
```

```python
df['pixel_density'] = ((df['resolution_width'] * df['resolution_height']) / 
                       (df['screen_size'] ** 2)).round(2)
```

```python
top_brands = df['brand_name'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 6))
top_brands.plot(kind='bar', color='skyblue')
plt.title("Top 10 Brands by Number of Smartphone Models")
plt.ylabel("Number of Models")
plt.xlabel("Brand")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_price = df.groupby('brand_name')['price'].mean().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
avg_price.plot(kind='bar', color='salmon')
plt.title("Top 10 Brands by Average Smartphone Price")
plt.ylabel("Average Price")
plt.xlabel("Brand")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
num_cols = df.select_dtypes(include='number')
plt.figure(figsize=(10, 8))
sns.heatmap(num_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['battery_capacity'], df['price'], alpha=0.5, color='blue')
plt.title("Price vs Battery Capacity")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df['month'] = pd.to_datetime(df['month'], errors='coerce')

tax_rate = 0.3

df['profit_before_tax'] = df['revenue'] - df['expenses']
df['tax_amount'] = df['profit_before_tax'] * tax_rate
df['PAT'] = df['profit_before_tax'] - df['tax_amount']

monthly_summary = df.groupby(df['month']).agg({
    'revenue': 'sum',
    'expenses': 'sum',
    'PAT': 'sum'
}).reset_index()

plt.figure(figsize=(20, 10))
plt.plot(monthly_summary['month'], monthly_summary['revenue'], marker='o', label='Revenue', color='blue')
plt.plot(monthly_summary['month'], monthly_summary['expenses'], marker='o', label='Expenses', color='red')
plt.plot(monthly_summary['month'], monthly_summary['PAT'], marker='o', label='PAT', color='green')

plt.title('Month-wise Revenue, Expenses & PAT')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
